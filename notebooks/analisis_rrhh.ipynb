```python
df = pd.read_csv("hr_limpio.csv", sep=";")
print("✅ Datos cargados:", df.shape)
print(df.columns.tolist())
```

```python
rotacion = df.groupby("Department")["Attrition"].apply(
    lambda x: (x == "Yes").sum() * 100 / len(x)).reset_index()
rotacion.columns = ["Departamento", "Porcentaje_Rotacion"]

plt.figure(figsize=(8,5))
sns.barplot(data=rotacion, x="Departamento", y="Porcentaje_Rotacion", palette="Reds_r")
plt.title("Tasa de Rotacion por Departamento", fontsize=15, fontweight="bold")
plt.ylabel("% Rotacion")
plt.xlabel("")
plt.tight_layout()
plt.savefig("grafico1_rotacion.png", dpi=150)
plt.show()
print("✅ Grafico 1 guardado")
```

```python
sueldo = df.groupby("JobRole")["MonthlyIncome"].mean().sort_values(ascending=True)

plt.figure(figsize=(10,6))
sueldo.plot(kind="barh", color="#1F5C99")
plt.title("Sueldo Promedio por Cargo", fontsize=15, fontweight="bold")
plt.xlabel("Ingreso Mensual Promedio")
plt.ylabel("")
plt.tight_layout()
plt.savefig("grafico2_sueldos.png", dpi=150)
plt.show()
print("✅ Grafico 2 guardado")
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x="Attrition", y="MonthlyIncome",
            palette={"Yes":"#E74C3C", "No":"#1F5C99"})
plt.title("Sueldo Mensual vs Rotacion", fontsize=15, fontweight="bold")
plt.xlabel("Renuncio?")
plt.ylabel("Ingreso Mensual")
plt.tight_layout()
plt.savefig("grafico3_sueldo_rotacion.png", dpi=150)
plt.show()
print("✅ Grafico 3 guardado")
```
